```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

### Exponentila Growth
$$\frac{dW}{dt} = a W$$

$$W(t) = W_0 e^{a t}$$

```python
# dW/dt for exponential growth
def exponential_growth(w,t):
    a = 0.05
    dwdt = a * w
    return dwdt

# Initial condition
w0 = 1

# time points
t = np.linspace(0,200)

# Solve the ODE
w_exponential = odeint(exponential_growth,w0,t)

# Plot W(t)
plt.plot(t,w_exponential, label='Exponential')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.legend()
plt.grid()
plt.show()
```

### Logistic growth
$$\frac{dW}{dt} = a W \left(1-\frac{W}{K}\right) $$

$$W(t) = \frac{K W_0 e^{a t}}{K+W_0(e^{a t}-1)}$$

```python
# dW/dt for logistic growth
def logistic_growth(w,t):
    a = 0.05
    k = 100
    dwdt = a * w * (1 - w/k)
    return dwdt

# initial condition
w0 = 1

# time points
t = np.linspace(0,200)

# solve ODE
w_logistic = odeint(logistic_growth,w0,t)

# plot results
plt.plot(t,w_logistic, label='Logistic')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.legend()
plt.grid()
plt.show()
```

### Comparison in the logarithmic scale

```python
plt.plot(t,w_exponential, label='Exponential')
plt.plot(t,w_logistic, label='Logistic')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```
