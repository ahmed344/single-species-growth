```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

### Exponentila growth
$$\frac{dW}{dt} = a W$$

$$W(t) = W_0 e^{a t}$$

```python
# dW/dt for exponential growth
def exponential_growth(w, t, a):
    dwdt = a * w
    return dwdt

# Parameters
a = 0.05

# Initial condition
w0 = 1

# Time points
t = np.linspace(0,200)

# Solve the ODE
w_exponential = odeint(exponential_growth, w0, t, args=(a,))

# Plot W(t)
plt.plot(t, w_exponential, label=f'Exponential a = {a}')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.title('Exponential gowth')
plt.legend(loc=(1.04,0))
plt.grid()
plt.show()
```

### Logistic growth
$$\frac{dW}{dt} = a W \left(1-\frac{W}{K}\right) $$

$$W(t) = \frac{K W_0 e^{a t}}{K+W_0(e^{a t}-1)}$$

```python
# dW/dt for logistic growth
def logistic_growth(w, t, a, k):
    dwdt = a * w * (1 - w/k)
    return dwdt

# Parameters
a = 0.05
k = 100

# Initial condition
w0 = 1

# Time points
t = np.linspace(0,200)

# solve ODE
w_logistic = odeint(logistic_growth, w0, t, args=(a, k))

# Plot W(t)
plt.plot(t, w_logistic, label=f'Logistic a = {a}')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.title('Logistic gowth')
plt.legend(loc=(1.04,0))
plt.grid()
plt.show()
```

### Comparison in the logarithmic scale

```python
plt.plot(t, w_exponential, label=f'Exponential a = {a}', alpha=0.8)
plt.plot(t, w_logistic, label=f'Logistic a = {a}, k = {k}', alpha=0.8)
plt.xlabel('t')
plt.ylabel('W(t)')
plt.yscale('log')
plt.title('Exponential vs Logistic gowth')
plt.legend(loc=(1.04,0))
plt.grid()
plt.show()
```

### Different growth rates

```python
growth_rates = [0.045, 0.05, 0.055]

# Plot W(t) for the exponential growth
for growth in growth_rates:
    plt.plot(t, odeint(exponential_growth, w0, t, args=(growth,)),
             label=f'Exponential a = {growth}', alpha=0.8)
plt.xlabel('t')
plt.ylabel('W(t)')
plt.title('Exponential gowth')
plt.legend(loc=(1.04,0))
plt.grid()
plt.show()

# Plot W(t) for the logitic growth
for growth in growth_rates:
    plt.plot(t, odeint(logistic_growth, w0, t, args=(growth, k)),
             label=f'Logistic a = {growth}, k = {k}', alpha=0.8)
plt.xlabel('t')
plt.ylabel('W(t)')
plt.title('Logistic gowth')
plt.legend(loc=(1.04,0))
plt.grid()
plt.show()

# Plot the comparison in logarithmic scale
for growth in growth_rates:
    plt.plot(t, odeint(exponential_growth, w0, t, args=(growth,)),
             label=f'Exponential a = {growth}', alpha=0.8)
for growth in growth_rates:
    plt.plot(t, odeint(logistic_growth, w0, t, args=(growth, k)),
             label=f'Logistic a = {growth}, k = {k}', alpha=0.8)
    
plt.xlabel('t')
plt.ylabel('W(t)')
plt.yscale('log')
plt.title('Exponential vs Logistic gowth')
plt.legend(loc=(1.04,0))
plt.grid()
plt.show()
```
